```python
#Treinando de visualização de dados
#Base de dados contem cod, nome, Pib, e empenho de munícipios
#A base já está devidamente limpa
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as srn
```

```python
dataset = pd.read_csv("dados/dados.csv", sep=";")

#Visualizando os dados
dataset.head()
```

```python
#LEMBRANDO, OS OUTLIERS SÃO DEFINIDOS PELA BIBLIOTECA
#Visualizando o PIB, com outliers

plt.boxplot(dataset.PIB, vert = False, showfliers = True, notch = True, patch_artist = True)
plt.title('PIB')
plt.xlabel('Quantidade(R$)')
```

```python
#Visualizando o PIB, sem outliers
#Parâmetro responsável por permite ou não o outlier é o showfilers
plt.boxplot(dataset.PIB, vert = False, showfliers = False, notch = True, patch_artist = True)
plt.title('PIB')
plt.xlabel('Quantidade(R$)')
```

```python
#Visualizando o Empenho, com outliers

plt.boxplot(dataset.VALOREMPENHO, vert = False, showfliers = True, notch = True, patch_artist = True)
plt.title('PIB')
plt.xlabel('Quantidade(R$)')
```

```python
#Visualizando o Empenho, SEM outliers

plt.boxplot(dataset.VALOREMPENHO, vert = False, showfliers = False, notch = True, patch_artist = True)
plt.title('PIB')
plt.xlabel('Quantidade(R$)')
```

```python
#Gerando um gráfico com os 10 maiores PIB's
agrupado = dataset.sort_values('PIB').head(10)
#Faço uma seleção de todas as colunas exceto a coluna de Empenho
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar( x='MUNICIPIO', y='PIB', color = 'gray')
```

```python
#Gerando um gráfico com os 10 maiores Empenhos's
agrupado = dataset.sort_values('VALOREMPENHO').head(10)
#Faço uma seleção de todas as colunas exceto a coluna de PIB
agrupado = agrupado.iloc[:,[1,3]]
agrupado.plot.bar( x='MUNICIPIO', y='VALOREMPENHO', color = 'gray')
```
